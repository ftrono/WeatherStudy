```python
import os
import numpy as np
import pandas as pd
```

Params

```python
DATASET = os.path.join("..", "datasets", "weather.csv")
```

Visualize full dataset

```python
data = pd.read_csv(DATASET)

data.drop(columns=["Unnamed: 0"], inplace=True)

data
```

**DF info & NaN analysis**

Show percentage of NaN values for every feature

```python
print(f"Total number of records: {len(data)}")
print(f"Total number of columns: {len(data.columns)}")
print(f"Total number of locations: {len(data["Location"].unique().tolist())}")

data_stats = pd.DataFrame(data={
    "DType": data.dtypes,
    "NaCount": len(data) - data.T.count(axis=1),
    "NaPct": round(((len(data) - data.T.count(axis=1)) / len(data)) * 100, 1),
    })
data_stats
```

Show percentage of NaN values for every feature and location

```python
loc_view = data.drop(columns='Location', axis=1).groupby(data['Location']).apply(lambda x: x.isna().sum()/len(x)*100).round(1)
loc_view
```

Add to DF info

```python
data_stats["LocsWhere>30%NA"] = loc_view.T.where(loc_view.T >= 30, np.nan).count(axis=1)
data_stats["LocsWhere100%NA"] = loc_view.T.where(loc_view.T == 100, np.nan).count(axis=1)
data_stats["LocsWhere>30%NA"] = data_stats["LocsWhere>30%NA"].fillna(0).astype(int)
data_stats["LocsWhere100%NA"] = data_stats["LocsWhere100%NA"].fillna(0).astype(int)
data_stats
```

Show what are the locations with more than 30% NaN values

```python
na_feats = data_stats[data_stats["NaPct"] > 30]
na_feats
```

```python
na_locs = []
for feat in na_feats.index:
    locs = loc_view[loc_view[feat] == 100].index.tolist()
    na_locs += locs
    print(f"{len(locs)}: {locs}")
```

```python
all_na_locs = set(na_locs)
print(f"Total locations with more than 30% NaN values: {len(all_na_locs)} / {len(data['Location'].unique().tolist())}")
all_na_locs
```

**Recap**

For 4 features out of the 23 columns, I have 22 locations out of 49 with more than 30% NaN data.

Since data are daily and highly variable, using interpolation through an Imputer would not make sense (I made an experiment anyway in the notebook "DecTree_imputer.ipynb"). Therefore, filling the gaps in the dataset would require to train a predictive model on the remaining data anyway.

I have two choices to clean the dataset:

* A) drop all rows with NaN, thus renouncing at data from these 22 locations and sparse missing values (note: temporal info is not needed in this dataset);
* B) drop the 4 features first, and then drop all rows with NaN.

Considering that the majority of NaN are concentrated in these 4 columns and that the 22 locations identified have enough data for all the remaining features, option B should be preferred.

To have an idea of the impact in terms of data loss:

```python
# OPTION A)
data_a = data.dropna().reset_index(drop=True)
print(f"Option A) Keep {len(data_a.index)} records out of {len(data.index)} ({round((len(data_a.index) / len(data.index))*100, 1)}%)")
```

```python
# OPTION B)
data_b = data.drop(columns=na_feats.index).dropna().reset_index(drop=True)
print(f"Option B) Keep {len(data_b.index)} records out of {len(data.index)} ({round((len(data_b.index) / len(data.index))*100, 1)}%)")
```

Definitely, option B is preferred.
