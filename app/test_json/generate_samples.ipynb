### Generate sample test JSON files

```python
import os, json
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
DATASET = os.path.join("datasets", "weather.csv")
TEST_PCT = 0.30
FEATURES_TO_REMOVE = ["Unnamed: 0"]
TARGET = "RainTomorrow"
MAX_FILES = 10
```

```python
data = pd.read_csv(DATASET)

data = data.drop(columns=FEATURES_TO_REMOVE).dropna().reset_index(drop=True)
data[TARGET] = np.where(data[TARGET] == "Yes", 1, 0)
data
```

```python
X = data.drop(columns=TARGET)
Y = data[TARGET]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=TEST_PCT, random_state=42)

X.shape, Y.shape, X_train.shape, Y_train.shape, X_test.shape, Y_test.shape
```

```python
X_test.reset_index(drop=True, inplace=True)
Y_test.reset_index(drop=True, inplace=True)
```

Export rows as JSON objects

```python
for i in range(len(X_test.index[:MAX_FILES])):
    temp = {"X": X_test.iloc[i].to_dict(), "Y": int(Y_test.iloc[i])}
    json.dump(temp, open(f"sample_{i+1}.json", "w"))
```
